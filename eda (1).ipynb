```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset


df =pd.read_csv('niftydata20yrs.csv')

# 1. Summary Statistics
summary_stats = df.describe()
summary_stats

```

```python
# 2. Data Visualization
# Plot time series graphs for each year
plt.figure(figsize=(12, 6))
for year in df['Year']:
    plt.plot(df.columns[1:], df[df['Year'] == year].values[0][1:], label=year)
plt.xlabel('Month')
plt.ylabel('nifty')
plt.title('Monthly nifty Variations Over Years')
plt.legend()
plt.show()

# Create box plots for each month
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop(columns='Year'))
plt.xlabel('Month')
plt.ylabel('nifty')
plt.title('Box Plot of Monthly  nifty  Distribution')
plt.show()

# Plot histograms for each month
plt.figure(figsize=(12, 6))
df.drop(columns='Year').hist(bins=20, figsize=(12, 10))
plt.suptitle('Histograms of Monthly nifty  Distribution', y=0.92)
plt.show()

# Use heatmap to visualize the correlation between different months
plt.figure(figsize=(10, 8))
sns.heatmap(df.drop(columns='Year').corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```

```python
# 9. Frequency Analysis (extreme weather events)
# Identify extreme weather events based on temperature thresholds
threshold_high = 10  # Example threshold for extremely high temperatures
threshold_low = -10  # Example threshold for extremely low temperatures

# Count occurrences of extreme events for each year
extreme_high_count = df[df.drop(columns='Year') > threshold_high].count(axis=1)
extreme_low_count = df[df.drop(columns='Year') < threshold_low].count(axis=1)

# Plot frequency of extreme weather events
plt.figure(figsize=(12, 6))
plt.plot(df['Year'], extreme_high_count, label='Extremely High (> 10)')
plt.plot(df['Year'], extreme_low_count, label='Extremely Low (< -10)')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Frequency Analysis of Extreme nifty')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 10. Time Series Decomposition (using seasonal decomposition of time series (STL) or other methods)
# Decompose the time series data
from statsmodels.tsa.seasonal import seasonal_decompose

# Select the time series to decompose (e.g., annual temperatures)
time_series = df['Annual']

# Perform time series decomposition
decomposition = seasonal_decompose(time_series, model='additive', period=12)

# Plot the decomposition components
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(decomposition.observed)
plt.title('Observed')

plt.subplot(4, 1, 2)
plt.plot(decomposition.trend)
plt.title('Trend')

plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal)
plt.title('Seasonal')

plt.subplot(4, 1, 4)
plt.plot(decomposition.resid)
plt.title('Residual')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a single figure to contain all scatter plots
fig, axes = plt.subplots(len(df.columns) - 2, 1, figsize=(10, 8 * (len(df.columns) - 2)))

# Iterate through all columns except 'Year' and 'Annual'
for i, column in enumerate(df.columns):
    if column not in ['Year', 'Annual']:
        axes[i - 1].scatter(df['Year'], df[column])
        axes[i - 1].set_xlabel('Year')
        axes[i - 1].set_ylabel(column)
        axes[i - 1].set_title(f'Scatter Plot of Year vs {column}')

# Adjust layout
plt.tight_layout()

# Save the figure as a JPEG file
plt.savefig('scatter_plots.jpg')
plt.show()
```
